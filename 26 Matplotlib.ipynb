### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

**Matplotlib** is a Python library used for creating static, animated, and interactive visualizations in Python. It is a popular data visualization library that provides a wide range of tools for creating high-quality 2D or 3D plots, graphs, and charts

Five plots that can be plotted using the Pyplot module of Matplotlib are:

1. Line plot: A line plot is used to visualize the relationship between two variables, usually time and some measurement.

2. Scatter plot: A scatter plot is used to visualize the relationship between two variables in a dataset. It is often used to identify correlations between variables.

3. Histogram: A histogram is a graphical representation of the distribution of a dataset. It shows the frequency distribution of a variable by dividing the data into bins and counting the number of observations in each bin.

4. Bar plot: A bar plot is used to compare different categories of data. It is useful for displaying the relative size of each category.

5. Box plot: A box plot is used to display the distribution of a dataset. It shows the median, quartiles, and outliers of a dataset, making it useful for identifying any skewness or outliers in the data

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Note: Also add title, xlabel, and ylabel to the plot.

**Scatter plot:** A scatter plot is used to visualize the relationship between two variables in a dataset. It is often used to identify correlations between variables.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.xlabel('this is my x - axis')
plt.ylabel('this is my y - axis')
plt.title("Scatter  Plot")
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:


For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import numpy as np


plt.rcParams["figure.autolayout"] = True
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title("For line 1")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title("For line 2")


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.title("For line 3")


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150]) 
plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.title("For line 4")

plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

**Bar plot:** A bar plot is used to compare different categories of data. It is useful for displaying the relative size of each category.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company, profit)
plt.show()
```

```python
plt.barh(company, profit)
plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

**Box plot:** A box plot is used to display the distribution of a dataset. It shows the median, quartiles, and outliers of a dataset, making it useful for identifying any skewness or outliers in the data

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2])
plt.show()
```
