## Q1. How can you create a Bokeh plot using Python code?


```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure , output_file, show
```

```python
x = [2,3,4,5,6]
y = [3,4,5,6,7]


output_file('line.html')
p = figure(title = 'line plot')
p.line(x,y)
show(p)
```

## Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.


Glyphs are the basic building blocks of Bokeh plots, representing the visual elements that are rendered on the plot, such as lines, circles, rectangles, and text.

```python
p.circle(x, y, legend_label="Circles", size=10, fill_color="white", line_color="blue")
show(p)
```

## Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?


```python

p.title.text = "Custom Title"
p.xaxis.axis_label = "Custom X Label"
p.yaxis.axis_label = "Custom Y Label"
```

```python
p.xaxis.major_label_orientation = "vertical"
```

```python

p.legend.location = "top_left"
p.legend.label_text_font_size = "12pt"
```

```python
p.circle(x, y, legend_label="Circles", size=10, fill_color="white", line_color="blue")
show(p)
```
